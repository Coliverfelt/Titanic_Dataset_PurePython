


### **Converting titanic_csv to a Dataset(dictionary format) using Python**




```python
# Reading lines
# Converting String to List
titanic_list = titanic_csv.readlines()
```

```python
type(titanic_list)
```

```python
print(titanic_list[0])
print(titanic_list[1])
```

```python
# Close the file to free memory
titanic_csv.close()
```

```python
# Setting the columns
columns = titanic_list[0].split(",")
print(type(columns))
print("\n", columns)
```

```python
del titanic_list[0]
```

```python
print(columns)
print("\n", titanic_list)
```

```python
# Creating a dictionary to organize data
dataset = {}

# Put columns as keys
for c in columns:
  dataset[c] = []

print(dataset)
```

```python
# Making Regex
def concName(linha):
  import re

  match = re.compile(r'"(.*)(,)(\s.*)"') # Compiling regex
  # (.*)   - Selecting all
  # (,)    - Selecting the comma
  # (/s.*) - Selecting all after the first space
  return match.sub(r'"\3 \1"', linha) # Sorting the line excluding the comma
```

```python
# Aplying Regex on titanic_list
for line in range(len(titanic_list)):
  titanic_list[line] = concName(titanic_list[line])

print(titanic_list)
```

```python
#Splitting data by comma
for line in range(len(titanic_list)):
  titanic_list[line] = titanic_list[line].split(",")
```

```python
print(titanic_list)
```

```python
# Setting data into the dataset
i = 0

for c in columns:
  for l in range(len(titanic_list)):
    dataset[c].append(titanic_list[l][i])
  i += 1
```

```python
print(dataset)
```

```python
for k, d in enumerate(dataset):
  print("{}: {}".format(d, dataset[d]))
```


### **Handling with Titanic Dataset(dictionary format) using Python**


**Filling Null values in dataset['Age'] with the average age**

```python
# Fill Null spaces in dataset['Age'] with -1
count1 = 0
for k, d in enumerate(dataset['Age']):
  if d == '':
    dataset['Age'][k] = -1
    count1 += 1
  else:
    dataset['Age'][k] = float(dataset['Age'][k])

print(dataset['Age'])
```

```python
# Length with (-1)
print(len(dataset['Age']))

# Length without (-1)
lengthage = len(dataset['Age']) - count1
print(lengthage)

# Sum with (-1)
print(sum(dataset['Age']))

# sum without (-1)
sumage = sum(dataset['Age']) + count1
print(sumage)
```

```python
# Age average
ageaverage = round(sumage/lengthage, 2)

print(ageaverage)
```

```python
# Filling dataset['Age'] with average
for d in range(len(dataset['Age'])):
  if dataset['Age'][d] == -1:
    dataset['Age'][d] = ageaverage

#for k, d in enumerate(dataset['Age']):
#    if d == -1:
#        dataset['Age'][k] = ageaverage

print(dataset['Age'])
```
